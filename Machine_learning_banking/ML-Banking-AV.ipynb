# Preparing the dataset for testing and training

```python
X_train = df.drop(columns = ["Interest_Rate"])
```

```python
Y_train = df.Interest_Rate
```

```python
X_train.Loan_Amount_Requested = X_train.Loan_Amount_Requested.apply(lambda x:x.replace(',',''))
```

```python
X_train.Loan_Amount_Requested = X_train.Loan_Amount_Requested.astype("float")
X_train.Length_Employed = X_train.Length_Employed.astype("category")
X_train.Income_Verified = X_train.Income_Verified.astype("category")
X_train.Home_Owner = X_train.Home_Owner.astype("category")
X_train.Purpose_Of_Loan = X_train.Purpose_Of_Loan.astype("category")
X_train.Inquiries_Last_6Mo = X_train.Inquiries_Last_6Mo.astype("category")
X_train.Gender = X_train.Gender.astype("category")
```

```python
X_train.info()
```

```python
columns = ["Length_Employed","Home_Owner","Income_Verified","Purpose_Of_Loan","Inquiries_Last_6Mo","Gender"]
```

```python
X_train = pd.get_dummies(X_train, columns=columns)
```

```python
X_train.shape
```

```python
X_train
```

```python
X_train["Loan_Amount_Bins"],bins_income = pd.qcut(X_train['Loan_Amount_Requested'],4,labels=['Loan_Category_{}'.format(i) for i in range(4)],retbins=True)
```

```python
X_train["Annual_Income_Bins"],bins = pd.qcut(X_train['Annual_Income'],4,labels=['Income_Category_{}'.format(i) for i in range(4)],retbins=True)
```

```python
X_train
```

```python
X_train = pd.get_dummies(X_train, columns=["Loan_Amount_Bins","Annual_Income_Bins"])
```

```python
X_train
```

```python
X_train1 = X_train
```

```python
X_train1 = X_train1.drop(columns=["Loan_Amount_Requested","Annual_Income"])
```

```python
X_train1.shape
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st = StandardScaler()
```

```python
st.fit(X_train)
X_train = st.transform(X_train)
```
